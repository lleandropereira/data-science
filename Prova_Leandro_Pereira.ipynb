```python
from scipy import stats

def testar_normalidade(dataset):
    resultados = {}

    for coluna in dataset.columns:
        variavel = dataset[coluna]

        if len(variavel) < 3:
            resultados[coluna] = 'Não suficiente'
        else:
            shapiro_results = stats.shapiro(variavel)
            p_valor = shapiro_results[1]

            if p_valor > 0.05:  # Usando um nível de significância de 0.05
                resultados[coluna] = 'Normal'
            else:
                resultados[coluna] = 'Não normal'

    return resultados

pares_de_indicadores = [
    ('Taxes on international trade (% of revenue)', 'Air transport, passengers carried'),
    ('Primary income on FDI, payments (current US$)', 'Consumer price index (2010 = 100)'),
    ('Children in employment, work only, female (% of female children in employment, ages 7-14)', 'Gross enrolment ratio, tertiary, male (%)'),
    ('Net financial flows, others (NFL, current US$)', 'Children in employment, study and work, female (% of female children in employment, ages 7-14)'),
    ('Primary income on FDI, payments (current US$)', 'PPG, bilateral (NTR, current US$)'),
    ('Children in employment, self-employed, male (% of male children in employment, ages 7-14)', 'PPG, commercial banks (NFL, current US$)'),
    ('Net intake rate to Grade 1 of primary education, male (%)', 'Net incurrence of liabilities, domestic (current LCU)'),
    ('Interest payments on external debt (% of GNI)', 'GNI, PPP (current international $)'),
    ('Lower secondary completion rate, male (%)', 'Technical cooperation grants (BoP, current US$)'),
    ('Wage and salaried workers, total (% of total employed)', 'Primary to secondary general education transition rate, female (%)')
]

paises_selecionados = ['Brazil', 'Argentina', 'Germany', 'France', 'China', 'Japan', 'United States', 'Canada']


# Loop para analisar os 10 pares de indicadores
for indicador1, indicador2 in pares_de_indicadores:
    # Filtrar o DataFrame para países e indicadores selecionados
    filtro = (df['CountryName'].isin(paises_selecionados)) & ((df['IndicatorName'] == indicador1) | (df['IndicatorName'] == indicador2))
    dados_selecionados = df[filtro]

    # Reorganizar os dados em um formato mais legível
    dados_pivot = dados_selecionados.pivot_table(index=['CountryName'], columns='IndicatorName', values='Value')

    # Exibir a tabela de comparação para o par de indicadores atual
    print(f'Comparação entre {indicador1} e {indicador2}:\n')
    print(dados_pivot)
    print('\n')

    # Testar a normalidade das variáveis em dados_pivot
    resultados_normalidade = testar_normalidade(dados_pivot)

    # Exibir os resultados do teste de normalidade
    print('Resultados do teste de normalidade:')
    for coluna, resultado in resultados_normalidade.items():
        print(f'{coluna}: {resultado}')
    print('\n')
```
